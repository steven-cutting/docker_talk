[**Project Gutenberg's The Return of Sherlock Holmes, by Arthur Conan Doyle**](http://www.gutenberg.org/ebooks/108)

```python
ASH = sc.textFile('108.txt.utf-8')
```

```python
PUNCTUATION = r"""[,-/\#\$\%\(\)\*\+\<\=\>\@\[\\\]\^\`\{\|\}\~ :?&;!'"]+"""
```

```python
counts = tlz.pipe(ASH.flatMap(lambda x: re.split(PUNCTUATION, x)),  # Split on spaces an punctuation
                  lambda x: x.filter(lambda x: x.strip()),  # Remove whitespace only tokens.
                  lambda x: x.filter(lambda x: len(x) >= 3),  # Remove tokens less than three characters.
                  lambda x: x.map(lambda x: x.lower()),  # Make all lowercase.
                  lambda x: x.map(lambda x: (x, 1)),  # Place each token in a tuple with the number one.
                  lambda x: x.reduceByKey(lambda x,y:x+y))  # Calculate frequencies.
```

```python
# Collects all of the results and sorts them based token count.
collect_n_sort = lambda rdd,r=True: sorted(rdd.collect(), key=lambda t: t[1], reverse=r)
```

```python
collect_n_sort(counts)
```

```python
collect_n_sort(counts.filter(lambda x: x[1] > 10), r=False)
```

```python
CHARCOUNTS = tlz.pipe(ASH.map(lambda x: x.lower()),  # Make all lowercase.
                      lambda x: x.flatMap(lambda x: x),  # Split words into individual characters.
                      lambda x: x.map(lambda x: (x, 1)),  # Place each token in a tuple with the number one.
                      lambda x: x.reduceByKey(lambda x,y:x+y))  # Calculate frequencies.
```

```python
collect_n_sort(CHARCOUNTS)
```
